```python
import pandas as pd 
import pymongo

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.fastfood_db
```

```python
# Load in the Yelp data from MongoDB

yelp_df = pd.DataFrame(list(db.yelp.find()))
yelp_df.head()
```

```python
# Load in the Kaggle data from MongoDB

kaggle_df = pd.DataFrame(list(db.kaggle.find()))
kaggle_df.head()
```

```python
# Round off the latitude and longitude values and combine them in a single column that we can try and merge our tables with it

yelp_df["lat_lon"] = yelp_df['latitude'].round(decimals=3).astype(str) + "|" + yelp_df['longitude'].round(decimals=3).astype(str)
yelp_df[['name', 'latitude', 'longitude', 'lat_lon']].head()
```

```python
# Round off the latitude and longitude values and combine them in a single column that we can try and merge our tables with it

kaggle_df["lat_lon"] = kaggle_df['latitude'].round(decimals=3).astype(str) + "|" + kaggle_df['longitude'].round(decimals=3).astype(str)
kaggle_df[['name', 'latitude', 'longitude', 'lat_lon']].head()
```

```python
# Merge the two dataframes using the newly created lat_lon column

merged_df = yelp_df[['name', 'latitude', 'longitude', 'lat_lon']].merge(kaggle_df[['name', 'latitude', 'longitude', 'lat_lon']], on='lat_lon')
merged_df
```

```python
# 449 rows means we have lost over half of our data with this merge!
# It looks like the Lat-Lon combination we were using is too specific - we may have to reduce the number of decimal places to reconcile the two datasets
# However adding another decimal place would likely result in a much higher rate of mistaken identity between nearby restraunts
# There are already a number of cases happening with the current, 3 decimal place version (see below at index 41, 42 for example)

merged_df.loc[merged_df['name_x'] != merged_df['name_y']]
```

```python
# We are going to attempt a different method of combining our two datasets
# When the Yelp data was generated, we recorded which row in the kaggle data was used to generate that API response as "kaggle_id"
# As all of the Yelp data (except for a handful of occasions where the API returned nothing at all) has an associated kaggle_id, this merge should result in significantly less lost data

new_merged_df = yelp_df[['_id', 'kaggle_id','name', 'latitude', 'longitude', 'lat_lon']].merge(kaggle_df[['id', 'name', 'latitude', 'longitude', 'lat_lon']], left_on='kaggle_id', right_on='id')
new_merged_df
```

```python
# 958 rows is substantially better! Of course, the Yelp API is not infallible - it simply returns the closest match
# For chain fast food restraunts, this generally seems to be another location of the same name
# We can check this hypothesis by checking entries where the names do not precisely match
# All of the mismatches seem to cosmetic differences (apostrophes, capitalizations, extra words etc)

new_merged_df.loc[new_merged_df['name_x'] != new_merged_df['name_y']]
```

```python
# It therefore seems that we should be more concerned about different locations
# Fortunately now we have merged on kaggle_id, we can be more generous now with our longitude and latitude match
# As two locations of the same brand are more likely to be more than a hundred meters apart than any two restraunts

new_merged_df["lat_lon_x"] = new_merged_df['latitude_x'].round(decimals=2).astype(str) + "|" + new_merged_df['longitude_x'].round(decimals=2).astype(str)
new_merged_df["lat_lon_y"] = new_merged_df['latitude_y'].round(decimals=2).astype(str) + "|" + new_merged_df['longitude_y'].round(decimals=2).astype(str)
new_merged_df.loc[new_merged_df['lat_lon_x'] == new_merged_df['lat_lon_y']]
```

```python
# 727 rows! Thats certainly an improvement. 
# There is certainly still a danger that we are confusing two very close locations of the same name
# We can try and perform another level of checks using street addresses
# Which we haven't used earlier due to numerous cosmetic differences (Avenue instead of Ave, periods after building numbers etc) 
# For now, lets clean up the extra columns and weird names and then save as a new collection to the database

filtered_df = new_merged_df.loc[new_merged_df['lat_lon_x'] == new_merged_df['lat_lon_y']]
filtered_df = filtered_df[['_id', 'kaggle_id', 'name_x', 'lat_lon_x']]
filtered_df = filtered_df.rename(columns={'_id': 'yelp_id',
                                          'name_x': 'restraunt name',
                                          'lat_lon_x': 'latitude|longitude'})
filtered_df.head()
```

```python
db.combined_dataset.drop()
filtered_df.index = filtered_df.index.astype(str)
filtered_df.apply(lambda row: db.combined_dataset.insert_one(row.to_dict()), axis=1)
```
